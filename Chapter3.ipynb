```python
from numpy import *
import math
import matplotlib.pyplot as plt

t = [pow(10, i) for i in range(3)]
a = [8*x for x in t]
b = [4*x*(log(x)) for x in t]
c = [2 * (x**2) for x in t] 
d = [(x**3) for x in t] 
e = [(2**x) for x in t] 


plt.plot(t, a, 'r') # plotting t, a separately 
plt.plot(t, b, 'b') # plotting t, b separately 
plt.plot(t, c, 'g') # plotting t, c separately 
plt.plot(t, d, 'y') # plotting t, c separately 
plt.plot(t, e, 'm') # plotting t, c separately 

plt.yscale('log')

plt.show()
```
