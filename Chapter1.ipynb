**R-1.3**

```python
def minmax(data):
  min = data[0]
  max = data[0]
  for i in range(len(data)):
    if i+1 < len(data):
      if data[i+1] < min:
        min = data[i+1]
      if data[i+1] > max:
        max = data[i+1]
  return(min, max)
```

```python
data = [3, 7, -5, 1, -9, 32, 65, -6, 0]
minmax(data)
```

```python
def sum_of_sqrs(x):
  if x >= 0 and type(x) is int:
    total = 0
    for i in range(x):
      total += i**2
    return total
  else:
    print("Variable x must be a positive integer")
```

```python
sum_of_sqrs(5)
```

**C-1.14**

```python
from math import factorial

def combination(n, r):
  return int(factorial(n)/(factorial(r)*(factorial(n - r))))

def odd_multipliers(data):
  i = 0
  for item in data:
    if item % 2 == 1:
      i+=1
  return combination(i, 2) 
```

```python
data = [2, 3, 5, 9, 4, 7]
odd_multipliers(data)
```

```python
from random import randint
def shuffle_data(data):
  for i in range(len(data)):
    random_indx = randint(0,4)
    data[random_indx], data[random_indx - 1] = data[random_indx - 1], data[random_indx]
  return data
```

```python
data = [1, 2, 3, 4, 5]
shuffle_data(data)
```

```python
def count_vowels(text):
  vowels = ['a', 'e', 'o', 'u', 'i']
  ctr = 0
  for letter in text:
    if letter in vowels:
      ctr += 1
  return ctr
```

**C-1.28**

```python
def calculate_p_norm(v, p=2):
  total = 0
  for i in v:
    total += i**p
  total = total ** (1/p)
  return total
```

```python
v = [3, 4]
print("Euclidean norm: ", calculate_p_norm(v))
print("7-norm: ", calculate_p_norm(v, 7))
```

```python
def make_change(given, charged):
  rest = given - charged
  if rest <= 0:
    return print("You gave less money than you are charged")
  bills_coins = [100, 20, 10, 5, 1, 0.25, 0.1, 0.05, 0.01]
  change = {}
  for money in bills_coins:
    div, rest = divmod(rest, money)
    if div > 0:
      change[money] = int(div)
  return change
```

```python
make_change(379.98, 100)
```

**P-1.35**

```python
def bday_paradox(n):
  comb_of_two = combination(n, 2)
  two_have_dffrnt_bday = 1-(1/365)
  prob_dffrnt_bday = two_have_dffrnt_bday ** comb_of_two
  prob_bday_same = 1 - prob_dffrnt_bday
  return print("When number of people in the room is {}, the probability is: {} ".format(n, prob_bday_same))
```

```python
for n in range(5, 100, 5):
  bday_paradox(n)
```

**P-1.36**

```python
def count_unique_words(text):
  word_list = text.split(" ")
  word_dict = {}

  for word in word_list:
    if word in word_dict.keys():
      word_dict[word] += 1
    else:
      word_dict[word] = 1
  return word_dict
```

```python
words = "jerry kramer newman george elaine jerry larry kramer"
count_unique_words(words)
```
